```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = (cv2.cvtColor(cv2.imread('datasets/images/graph.png'), cv2.COLOR_BGR2GRAY) < 240).astype('uint8') * 255
kernel = np.ones((2, 2), np.uint8)
img = cv2.erode(img, kernel, iterations=1)
plt.imshow(img, cmap="gray")
plt.show()
edges = cv2.Canny(img, 252, 254)
plt.imshow(edges)
plt.show()

# TODO: restore the nodes and the edges or the graph by picture
```
