```python
# please refer to source: https://medium.com/@sebastiannorena/finding-correlation-between-many-variables-multidimensional-dataset-with-python-5deb3f39ffb3

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('datasets/ml/correlation.csv', index_col=0)
corr = data.corr()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
# cax = ax.matshow(abs(corr), cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, len(data.columns), 1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns)
ax.set_yticklabels(data.columns)
plt.show()

corr.style.background_gradient()
```

```python
print(data.var())
print("----------------")
print(np.var(data['delta'].values))

vector = data['delta'].values
avg = (vector.sum() / vector.shape[0])
squares = ((vector - avg) ** 2).sum()

print(squares / vector.shape[0])
```

```python
from random import random
from sklearn.decomposition import PCA

X = np.matrix([[x + random() - 0.5, x + random() - 0.5] for x in range(-10, 10)])

def plot(X):
    plt.plot(X[:,0], X[:,1])
    plt.show()
    print("X variance = {}, Y variance = {}".format(X[:,0].var(), X[:,1].var()))

plot(X)

pca = PCA(n_components=2)
X_ = pca.fit_transform(X)

plot(X_)
```
