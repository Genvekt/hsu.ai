```python
import itertools
import copy

class Field(object):
    state = None
    parent = None
    children = []
    minimax_score = None
    
    def __init__(self, state):
        self.state = state
        self.children = []
    
    def print_me(self):
        for line in self.state:
            print(*line, sep='')
    
    def is_move_of(self, side):
        my_moves = 0
        enemy_moves = 0
        for row in range(3):
            for column in range(3):
                c = self.state[row][column]
                if c == side:
                    my_moves += 1
                elif c == ('x' if (side == 'o') else 'o'):
                    enemy_moves += 1
        
        if side == 'x':
            return my_moves == enemy_moves
        else:
            return my_moves < enemy_moves

    def is_win_of(self, side):
        for row in range(3):
            if self.state[row][0] == side \
                and self.state[row][0] == self.state[row][1] \
                and self.state[row][0] == self.state[row][2]:
                return True
        for column in range(3):
            if self.state[0][column] == side \
                and self.state[0][column] == self.state[1][column] \
                and self.state[0][column] == self.state[2][column]:
                return True
        
        if self.state[0][0] == side \
            and self.state[0][0] == self.state[1][1] \
            and self.state[0][0] == self.state[2][2]:
            return True
        
        if self.state[0][2] == side \
            and self.state[0][2] == self.state[1][1] \
            and self.state[0][2] == self.state[2][0]:
            return True
        
        return False
    
    def get_next_move(self, best_for_side):
        # return field, which is best in terms of minimax
        return children[-1]
```

```python
def generate_children(field):
    # TODO: write code that generates children fields.
    # actually, make moves where _possible_
    result = 0
    side = 'x'
    if field.is_move_of('o'):
        side = 'o'
        
    if side == 'o' and field.is_win_of('x'):
        return result
    if side == 'x' and field.is_win_of('o'):
        return result
    
#     print(field.state)
    for row in range(3):
        for column in range(3):
            if (field.state[row][column] == '.'):
                field.state[row][column] = side
#                 print(field.state)
                
                one_field = Field(copy.deepcopy(field.state))
                field.children.append(one_field)
#                 print(one_field.state)
                field.state[row][column] = '.'
    if field.children is not None:
        result = len(field.children)
    for child in field.children:
        result += generate_children(child)
    
    
    return result


def update_minimax(field, side):
    #TODO write fuction that updates minimax weights for all nodes RECURSIVELY
    if side == 'o' and field.is_win_of('x'):
        field.minimax_score = 10
    elif side == 'x' and field.is_win_of('o'):
        field.minimax_score = -10
    elif side == 'x':
        minimum = 1000000
        for child in field.children:
            update_minimax(child, 'o')
            if minimum > child.minimax_score:
                minimum = child.minimax_score
        field.minimax_score = minimum
    elif side == 'o':
        maximum = -1000000
        for child in field.children:
            update_minimax(child, 'x')
            if maximum < child.minimax_score:
                maximum = child.minimax_score
        field.minimax_score = maximum
    
    return field.minimax_score
```
